```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
number_cat = 10
```

```python
y_train
```

```python
import keras
y_train = keras.utils.to_categorical(y_train,number_cat)
y_train
```

```python
y_test = keras.utils.to_categorical(y_test, number_cat)
```

```python
y_test
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train.shape
```

```python
Input_shape = X_train.shape[1:]
```

```python
Input_shape
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
from keras import backend as K
K.common.set_image_dim_ordering('th')
```

```python
cnn_model = Sequential()
cnn_model.add(Conv2D(filters = 1024, kernel_size = (3,3), activation = 'relu', input_shape = Input_shape,padding='same'))
cnn_model.add(Conv2D(filters = 1024, kernel_size = (3,3), activation = 'relu',padding='same'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.5))


cnn_model.add(Conv2D(filters = 1024, kernel_size = (3,3), activation = 'relu',padding='same'))
cnn_model.add(Conv2D(filters = 1024, kernel_size = (3,3), activation = 'relu',padding='same'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.5))


cnn_model.add(Conv2D(filters = 1024, kernel_size = (3,3), activation = 'relu',padding='same'))
cnn_model.add(Conv2D(filters = 1024, kernel_size = (3,3), activation = 'relu',padding='same'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.5))

cnn_model.add(Flatten())

cnn_model.add(Dense(units = 2048, activation = 'relu'))
cnn_model.add(Dense(units = 2048, activation = 'relu'))

cnn_model.add(Dense(units = 2048, activation = 'relu'))

cnn_model.add(Dense(units = 2048, activation = 'relu'))

cnn_model.add(Dense(units = 10, activation = 'softmax'))
```

```python
cnn_model.compile(loss = 'categorical_crossentropy', optimizer = keras.optimizers.rmsprop(lr = 0.001), metrics = ['accuracy'])
```

```python
history = cnn_model.fit(X_train, y_train, batch_size = 256, epochs = 50, shuffle = True)
```

```python
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Accuracy: {}'.format(evaluation[1]))
```

```python
predicted_classes = cnn_model.predict_classes(X_test) 
predicted_classes
```
